# Cleaning 'ingredients'

```python
# üîπ ‡∏•‡∏ö‡πÄ‡∏Ñ‡∏£‡∏∑‡πà‡∏≠‡∏á‡∏´‡∏°‡∏≤‡∏¢ * ‡∏ó‡∏∏‡∏Å‡πÅ‡∏ñ‡∏ß‡πÉ‡∏ô‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå ingredients ‡∏Å‡πà‡∏≠‡∏ô‡∏ï‡∏£‡∏ß‡∏à
df['ingredients'] = df['ingredients'].astype(str).str.replace('*', '', regex=False)
```

```python
df.loc[5, "ingredients"]
```

```python
# ‡∏û‡∏¥‡∏™‡∏π‡∏à‡∏ô‡πå‡∏ß‡πà‡∏≤ ingredients ‡∏°‡∏µ‡∏Ñ‡∏≥‡∏ã‡πâ‡∏≥
def has_duplicate_ingredients(text):
    parts = [p.strip().lower() for p in str(text).split(",")]
    return len(parts) != len(set(parts))

# ‡∏ñ‡πâ‡∏≤‡∏°‡∏µ‡∏Ñ‡πà‡∏≤ True ‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏°‡∏≤‡∏Å ‚Üí ‡πÅ‡∏õ‡∏•‡∏ß‡πà‡∏≤ ‚Äú‡∏°‡∏µ ingredients ‡∏ã‡πâ‡∏≥‡πÉ‡∏ô‡πÅ‡∏ñ‡∏ß‚Äù ‡∏à‡∏£‡∏¥‡∏á
df['has_dup_ing'] = df['ingredients'].apply(has_duplicate_ingredients)
df['has_dup_ing'].value_counts()
```

```python
# ‡∏£‡∏≤‡∏¢‡∏Å‡∏≤‡∏£‡πÅ‡∏ñ‡∏ß‡∏ó‡∏µ‡πà ingredients ‡∏ã‡πâ‡∏≥
dup_ing_df = df[df['has_dup_ing'] == True][['brand','name','ingredients']]
dup_ing_df.head(10)
```

```python
# "‡∏Ñ‡∏≥‡∏ó‡∏µ‡πà‡∏ã‡πâ‡∏≥" ‡πÉ‡∏ô‡πÅ‡∏ï‡πà‡∏•‡∏∞‡πÅ‡∏ñ‡∏ß
def find_duplicates(text):
    parts = [p.strip().lower() for p in str(text).split(",")]
    dups = [p for p in parts if parts.count(p) > 1]
    return list(set(dups))

df['dup_list'] = df['ingredients'].apply(find_duplicates)
df[df['has_dup_ing'] == True][['brand','name','dup_list']].head(10)
```

```python
# ‚Äú‡∏•‡πâ‡∏≤‡∏á‡∏ã‡πâ‡∏≥‚Äù ‡πÉ‡∏ô‡∏Ç‡∏±‡πâ‡∏ô‡∏ï‡∏≠‡∏ô Data Cleaning
def clean_ingredients(text):
    txt = str(text).lower().replace("‚Äô", "'")
    parts = [p.strip(" .") for p in txt.split(",")]
    seen, result = set(), []
    for p in parts:
        if p and p not in seen:
            seen.add(p)
            result.append(p)
    return " | ".join(result)

df["ingredients_list"] = df["ingredients"].apply(clean_ingredients)
```

```python
def has_duplicate_after_clean(text):
    parts = [p.strip() for p in str(text).split('|')]
    return len(parts) != len(set(parts))

df['dup_after_clean'] = df['ingredients_list'].apply(has_duplicate_after_clean)
df['dup_after_clean'].value_counts()
```

```python
def extract_ingredients_only(text: str) -> str:
    s = str(text)

    # 1) ‡∏£‡∏ß‡∏°‡∏ö‡∏£‡∏£‡∏ó‡∏±‡∏î‡πÉ‡∏´‡πâ‡πÄ‡∏õ‡πá‡∏ô‡∏ö‡∏£‡∏£‡∏ó‡∏±‡∏î‡πÄ‡∏î‡∏µ‡∏¢‡∏ß (‡∏Å‡∏±‡∏ô \r \n ‡πÅ‡∏õ‡∏•‡∏Å ‡πÜ)
    s = re.sub(r'[\r\n]+', ' ', s)

    # 2) ‡∏´‡∏≤ keyword "ingredient / ingredients" ‡∏ï‡∏≤‡∏°‡∏î‡πâ‡∏ß‡∏¢ :, -, ‚Äì ‡πÅ‡∏•‡πâ‡∏ß‡∏ï‡∏±‡∏î‡∏™‡πà‡∏ß‡∏ô‡∏´‡∏ô‡πâ‡∏≤‡∏≠‡∏≠‡∏Å
    #    ‡πÉ‡∏ä‡πâ finditer ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡πÄ‡∏•‡∏∑‡∏≠‡∏Å "‡∏ï‡∏≥‡πÅ‡∏´‡∏ô‡πà‡∏á‡∏™‡∏∏‡∏î‡∏ó‡πâ‡∏≤‡∏¢" ‡πÄ‡∏ú‡∏∑‡πà‡∏≠‡∏°‡∏µ‡∏Ñ‡∏≥‡∏ô‡∏µ‡πâ‡∏ã‡πâ‡∏≥
    pat = re.compile(r'ingredient[s]?\s*[:\-‚Äì]\s*', flags=re.IGNORECASE)
    matches = list(pat.finditer(s))
    if matches:
        s = s[matches[-1].end():]  # ‡∏ï‡∏±‡∏î‡∏ó‡∏∏‡∏Å‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏Å‡πà‡∏≠‡∏ô "ingredients:" ‡∏≠‡∏≠‡∏Å

    # 3) ‡∏ó‡∏≥‡∏Ñ‡∏ß‡∏≤‡∏°‡∏™‡∏∞‡∏≠‡∏≤‡∏î‡∏™‡πÄ‡∏õ‡∏ã/‡∏Ñ‡∏≠‡∏°‡∏°‡∏≤‡πÄ‡∏•‡πá‡∏Å‡∏ô‡πâ‡∏≠‡∏¢
    s = re.sub(r'\s*,\s*', ', ', s)   # ‡∏ä‡πà‡∏≠‡∏á‡∏ß‡πà‡∏≤‡∏á‡∏£‡∏≠‡∏ö‡∏Ñ‡∏≠‡∏°‡∏°‡∏≤‡πÉ‡∏´‡πâ‡∏°‡∏≤‡∏ï‡∏£‡∏ê‡∏≤‡∏ô
    s = re.sub(r'\s+', ' ', s).strip()

    return s

# ‡πÉ‡∏ä‡πâ‡∏á‡∏≤‡∏ô‡∏Å‡∏±‡∏ö DataFrame ‡∏Ç‡∏≠‡∏á‡∏Ñ‡∏∏‡∏ì
df['ingredients_cleaned'] = df['ingredients'].apply(extract_ingredients_only)

# (‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏∑‡∏≠‡∏Å) ‡∏î‡∏π‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡πÅ‡∏ñ‡∏ß‡∏ó‡∏µ‡πà‡∏ñ‡∏π‡∏Å‡∏ï‡∏±‡∏î prefix ‡∏≠‡∏≠‡∏Å‡∏à‡∏£‡∏¥‡∏á ‡πÜ
num_changed = (df['ingredients'] != df['ingredients_cleaned']).sum()
print(f"‡∏ï‡∏±‡∏î‡∏Ñ‡∏≥‡∏≠‡∏ò‡∏¥‡∏ö‡∏≤‡∏¢‡∏´‡∏ô‡πâ‡∏≤ 'Ingredients:' ‡∏≠‡∏≠‡∏Å‡πÅ‡∏•‡πâ‡∏ß {num_changed} ‡πÅ‡∏ñ‡∏ß")

# (‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏∑‡∏≠‡∏Å) ‡∏î‡∏π‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏Å‡πà‡∏≠‡∏ô-‡∏´‡∏•‡∏±‡∏á
# df.loc[row_index, ['ingredients','ingredients_cleaned']]
```

```python

def check_invalid_reason(text):
    raw = str(text)
    t = raw.strip().lower()

    # ‡∏£‡∏ß‡∏°‡∏ö‡∏£‡∏£‡∏ó‡∏±‡∏î‡πÉ‡∏´‡πâ‡πÄ‡∏õ‡πá‡∏ô‡∏£‡∏π‡∏õ‡πÅ‡∏ö‡∏ö‡πÄ‡∏î‡∏µ‡∏¢‡∏ß
    t_nl = re.sub(r'\r\n?|\u2028|\u2029', '\n', t)

    # 1Ô∏è‚É£ ‡πÄ‡∏Ñ‡∏™‡πÅ‡∏ö‡∏ö bullet/‡∏´‡∏±‡∏ß‡∏Ç‡πâ‡∏≠ (‡πÄ‡∏ä‡πà‡∏ô "-Vitamin E: Acts as ...")
    bullet_lines = re.findall(
        r'^\s*[-‚Ä¢¬∑]?\s*[^:\n]{1,80}:\s*[^\n,]+',  # ‡∏ä‡∏∑‡πà‡∏≠‡∏´‡∏±‡∏ß‡∏Ç‡πâ‡∏≠: ‡∏ï‡∏≤‡∏°‡∏î‡πâ‡∏ß‡∏¢‡∏Ñ‡∏≥‡∏≠‡∏ò‡∏¥‡∏ö‡∏≤‡∏¢
        t_nl,
        flags=re.MULTILINE
    )
    if len(bullet_lines) >= 1:
        return f"‚ö†Ô∏è ‡∏Ç‡πâ‡∏≠‡∏Ñ‡∏ß‡∏≤‡∏°‡πÅ‡∏ö‡∏ö‡∏ö‡∏π‡∏•‡πÄ‡∏•‡πá‡∏ï/‡∏´‡∏±‡∏ß‡∏Ç‡πâ‡∏≠ {len(bullet_lines)} ‡∏ö‡∏£‡∏£‡∏ó‡∏±‡∏î‡∏ó‡∏µ‡πà‡∏°‡∏µ ':' (‡∏Ñ‡∏≥‡∏≠‡∏ò‡∏¥‡∏ö‡∏≤‡∏¢ ‡πÑ‡∏°‡πà‡πÉ‡∏ä‡πà‡∏™‡∏π‡∏ï‡∏£)"

    # 2Ô∏è‚É£ ‡∏ñ‡πâ‡∏≤‡∏°‡∏µ '-' ‡∏ó‡∏µ‡πà‡∏ï‡πâ‡∏ô‡∏ö‡∏£‡∏£‡∏ó‡∏±‡∏î (‡πÅ‡∏°‡πâ‡πÑ‡∏°‡πà‡∏°‡∏µ ':' ‡∏Å‡πá‡∏ñ‡∏∑‡∏≠‡∏ß‡πà‡∏≤‡∏ô‡πà‡∏≤‡∏™‡∏á‡∏™‡∏±‡∏¢)
    if re.search(r'^\s*-\s*[A-Za-z]', t_nl, flags=re.MULTILINE):
        return "‚ö†Ô∏è ‡∏°‡∏µ‡πÄ‡∏Ñ‡∏£‡∏∑‡πà‡∏≠‡∏á‡∏´‡∏°‡∏≤‡∏¢ '-' ‡∏ô‡∏≥‡∏´‡∏ô‡πâ‡∏≤‡∏ö‡∏£‡∏£‡∏ó‡∏±‡∏î (‡∏≠‡∏≤‡∏à‡πÄ‡∏õ‡πá‡∏ô‡∏Ñ‡∏≥‡∏≠‡∏ò‡∏¥‡∏ö‡∏≤‡∏¢ ‡πÑ‡∏°‡πà‡πÉ‡∏ä‡πà‡∏™‡∏π‡∏ï‡∏£‡∏à‡∏£‡∏¥‡∏á)"

    # 3Ô∏è‚É£ ‡∏Ç‡πâ‡∏≠‡∏Ñ‡∏ß‡∏≤‡∏°‡πÇ‡∏Ü‡∏©‡∏ì‡∏≤/‡∏ö‡∏£‡∏£‡∏¢‡∏≤‡∏¢‡∏ó‡∏±‡πà‡∏ß‡πÑ‡∏õ
    if re.search(r'visit the|provides|acts as|known to|powerful|boutique|rich in|treat|heal', t):
        return "‚ö†Ô∏è ‡∏°‡∏µ‡∏Ç‡πâ‡∏≠‡∏Ñ‡∏ß‡∏≤‡∏°‡πÇ‡∏Ü‡∏©‡∏ì‡∏≤‡∏´‡∏£‡∏∑‡∏≠‡∏ö‡∏£‡∏£‡∏¢‡∏≤‡∏¢ (e.g., provides / acts as / known to)"

    # 4Ô∏è‚É£ ‡∏™‡∏±‡∏ç‡∏•‡∏±‡∏Å‡∏©‡∏ì‡πå‡∏Å‡∏≤‡∏£‡∏ï‡∏•‡∏≤‡∏î
    if re.search(r'\*|\#|%|100%', t):
        return "‚ö†Ô∏è ‡∏°‡∏µ‡∏™‡∏±‡∏ç‡∏•‡∏±‡∏Å‡∏©‡∏ì‡πå‡∏´‡∏£‡∏∑‡∏≠‡∏Ñ‡∏≥‡∏ó‡∏≤‡∏á‡∏Å‡∏≤‡∏£‡∏ï‡∏•‡∏≤‡∏î (*, %, 100%)"

    # 5Ô∏è‚É£ ‡∏°‡∏µ ':' ‡∏°‡∏≤‡∏Å‡∏Å‡∏ß‡πà‡∏≤ 1 ‡∏ï‡∏±‡∏ß (backup)
    if t.count(':') > 1:
        return "‚ö†Ô∏è ‡∏°‡∏µ‡πÄ‡∏Ñ‡∏£‡∏∑‡πà‡∏≠‡∏á‡∏´‡∏°‡∏≤‡∏¢ ':' ‡∏°‡∏≤‡∏Å‡∏Å‡∏ß‡πà‡∏≤ 1 ‡∏ï‡∏±‡∏ß (‡∏≠‡∏≤‡∏à‡∏°‡∏µ‡∏´‡∏•‡∏≤‡∏¢‡∏™‡∏π‡∏ï‡∏£‡∏´‡∏£‡∏∑‡∏≠‡∏Ñ‡∏≥‡∏≠‡∏ò‡∏¥‡∏ö‡∏≤‡∏¢‡πÉ‡∏ô cell ‡πÄ‡∏î‡∏µ‡∏¢‡∏ß)"

    # 6Ô∏è‚É£ ‡∏™‡∏π‡∏ï‡∏£ Excel / #NAME?
    if raw.startswith('=') or '#name?' in t:
        return "‚ö†Ô∏è ‡∏Ç‡πâ‡∏≠‡∏Ñ‡∏ß‡∏≤‡∏°‡∏°‡∏µ‡∏•‡∏±‡∏Å‡∏©‡∏ì‡∏∞‡∏™‡∏π‡∏ï‡∏£ Excel (#NAME?, =...)"

    # 7Ô∏è‚É£ ‡∏™‡∏±‡πâ‡∏ô‡∏ú‡∏¥‡∏î‡∏õ‡∏Å‡∏ï‡∏¥
    if len(t.split()) < 5:
        return "‚ö†Ô∏è ‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏Ñ‡∏≥ ‡∏ô‡πâ‡∏≠‡∏¢‡πÄ‡∏Å‡∏¥‡∏ô‡πÑ‡∏õ ‚Äî ‡∏≠‡∏≤‡∏à‡πÑ‡∏°‡πà‡πÉ‡∏ä‡πà‡∏™‡∏π‡∏ï‡∏£ ingredients ‡∏à‡∏£‡∏¥‡∏á"

    # 8Ô∏è‚É£ ‡∏ï‡∏±‡∏ß‡∏≠‡∏±‡∏Å‡∏©‡∏£‡∏Ñ‡∏ß‡∏ö‡∏Ñ‡∏∏‡∏°‡∏´‡∏£‡∏∑‡∏≠ newline ‡πÅ‡∏õ‡∏•‡∏Å
    if re.search(r'[\x00-\x08\x0b-\x1f]+', raw):
        return "‚ö†Ô∏è ‡∏°‡∏µ‡∏ï‡∏±‡∏ß‡∏≠‡∏±‡∏Å‡∏©‡∏£‡∏Ñ‡∏ß‡∏ö‡∏Ñ‡∏∏‡∏°‡∏ã‡πà‡∏≠‡∏ô‡∏≠‡∏¢‡∏π‡πà"

    return ""
```

```python
# ‡πÄ‡∏û‡∏¥‡πà‡∏°‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå reason ‡πÅ‡∏™‡∏î‡∏á‡∏™‡∏≤‡πÄ‡∏´‡∏ï‡∏∏
df['invalid_reason'] = df['ingredients'].apply(check_invalid_reason)

# flag ‡∏ß‡πà‡∏≤‡∏ô‡πà‡∏≤‡∏™‡∏á‡∏™‡∏±‡∏¢‡πÑ‡∏´‡∏° (True/False)
df['invalid_ing'] = df['invalid_reason'] != ""

# ‡∏î‡∏π‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á‡πÅ‡∏ñ‡∏ß‡∏ó‡∏µ‡πà‡∏ô‡πà‡∏≤‡∏™‡∏á‡∏™‡∏±‡∏¢
invalid_df = df[df['invalid_ing'] == True][['brand', 'name', 'ingredients', 'invalid_reason']]
invalid_df.head(10)
```

# Merge 'ingredients' column

```python
def normalize_ing_string(s: str) -> str | pd._libs.missing.NAType:
    """
    ‡∏£‡∏±‡∏ö‡∏™‡∏ï‡∏£‡∏¥‡∏á‡∏™‡∏π‡∏ï‡∏£ ‚Üí ‡∏Ñ‡∏∑‡∏ô‡∏™‡∏ï‡∏£‡∏¥‡∏á‡∏ó‡∏µ‡πà‡∏™‡∏∞‡∏≠‡∏≤‡∏î:
      - ‡∏ï‡∏±‡∏î‡∏ä‡πà‡∏≠‡∏á‡∏ß‡πà‡∏≤‡∏á‡πÄ‡∏Å‡∏¥‡∏ô
      - split ‡∏î‡πâ‡∏ß‡∏¢ ',', strip ‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏ï‡∏±‡∏ß
      - ‡∏•‡∏ö‡∏ß‡πà‡∏≤‡∏á‡πÄ‡∏õ‡∏•‡πà‡∏≤
      - ‡∏•‡∏ö‡∏Ñ‡∏≥‡∏ã‡πâ‡∏≥‡πÇ‡∏î‡∏¢‡∏Ñ‡∏á‡∏•‡∏≥‡∏î‡∏±‡∏ö‡πÄ‡∏î‡∏¥‡∏°
      - join ‡∏Å‡∏•‡∏±‡∏ö‡∏î‡πâ‡∏ß‡∏¢ ', '
    """
    if pd.isna(s):
        return pd.NA
    s = str(s)
    # normalize space
    s = re.sub(r'\s+', ' ', s).strip()
    # tokenize
    parts = [p.strip() for p in s.split(',')]
    parts = [p for p in parts if p]  # ‡∏ï‡∏±‡∏î‡∏ß‡πà‡∏≤‡∏á
    # ‡∏•‡∏ö‡∏ã‡πâ‡∏≥‡πÅ‡∏ö‡∏ö‡∏Ñ‡∏á‡∏•‡∏≥‡∏î‡∏±‡∏ö
    seen = set()
    dedup = []
    for p in parts:
        if p not in seen:
            seen.add(p)
            dedup.append(p)
    return ', '.join(dedup) if dedup else pd.NA

def list_to_ing_string(x) -> str | pd._libs.missing.NAType:
    """
    ‡∏£‡∏±‡∏ö list ‡∏Ç‡∏≠‡∏á ingredients ‚Üí ‡∏Ñ‡∏∑‡∏ô‡∏™‡∏ï‡∏£‡∏¥‡∏á‡∏ó‡∏µ‡πà‡∏™‡∏∞‡∏≠‡∏≤‡∏î (dedup + join)
    """
    if isinstance(x, list):
        # ‡πÅ‡∏õ‡∏•‡∏á‡πÄ‡∏õ‡πá‡∏ô‡∏™‡∏ï‡∏£‡∏¥‡∏á‡∏Å‡πà‡∏≠‡∏ô‡πÅ‡∏•‡πâ‡∏ß normalize ‡∏ï‡πà‡∏≠
        return normalize_ing_string(', '.join(map(str, x)))
    # ‡∏ñ‡πâ‡∏≤‡πÑ‡∏°‡πà‡πÉ‡∏ä‡πà list (‡πÄ‡∏ä‡πà‡∏ô ‡πÄ‡∏õ‡πá‡∏ô NaN ‡∏´‡∏£‡∏∑‡∏≠‡πÄ‡∏õ‡πá‡∏ô‡∏™‡∏ï‡∏£‡∏¥‡∏á‡∏≠‡∏¢‡∏π‡πà‡πÅ‡∏•‡πâ‡∏ß) ‡∏Å‡πá normalize ‡πÅ‡∏ö‡∏ö‡∏™‡∏ï‡∏£‡∏¥‡∏á
    return normalize_ing_string(x)

# 1) ‡πÄ‡∏ï‡∏£‡∏µ‡∏¢‡∏° candidate ‡∏à‡∏≤‡∏Å ingredients_list (‡∏ñ‡πâ‡∏≤‡∏°‡∏µ)
if 'ingredients_list' in df.columns:
    cand_from_list = df['ingredients_list'].apply(list_to_ing_string)
else:
    cand_from_list = pd.Series(pd.NA, index=df.index)

# 2) ‡πÄ‡∏ï‡∏£‡∏µ‡∏¢‡∏° candidate ‡∏à‡∏≤‡∏Å ingredients_cleaned (‡∏ñ‡πâ‡∏≤‡∏°‡∏µ)
if 'ingredients_cleaned' in df.columns:
    cand_from_cleaned = df['ingredients_cleaned'].apply(normalize_ing_string)
else:
    cand_from_cleaned = pd.Series(pd.NA, index=df.index)

# 3) ‡∏£‡∏ß‡∏°‡πÄ‡∏õ‡πá‡∏ô‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡πÅ‡∏°‡πà 'ingredients'
#    ‡πÄ‡∏•‡∏∑‡∏≠‡∏Å‡∏Ñ‡πà‡∏≤‡∏à‡∏≤‡∏Å ingredients_list ‡∏Å‡πà‡∏≠‡∏ô (‡πÄ‡∏û‡∏£‡∏≤‡∏∞‡∏•‡∏ö‡∏ã‡πâ‡∏≥‡πÅ‡∏•‡πâ‡∏ß)
#    ‡∏ñ‡πâ‡∏≤‡∏ß‡πà‡∏≤‡∏á‡∏Ñ‡πà‡∏≠‡∏¢ fallback ‡πÑ‡∏õ‡πÉ‡∏ä‡πâ ingredients_cleaned
ingredients_merged = cand_from_list.fillna(cand_from_cleaned)

# (‡∏≠‡∏≠‡∏õ‡∏ä‡∏±‡∏ô) ‡∏ñ‡πâ‡∏≤‡∏Ñ‡∏∏‡∏ì‡∏ï‡πâ‡∏≠‡∏á‡∏Å‡∏≤‡∏£ ‚Äú‡∏Ñ‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡πÅ‡∏ñ‡∏ß‡∏ó‡∏µ‡πà‡∏ô‡πà‡∏≤‡∏™‡∏á‡∏™‡∏±‡∏¢‡πÑ‡∏ß‡πâ‚Äù ‡∏Å‡πá‡πÄ‡∏ã‡πá‡∏ï‡∏ï‡∏£‡∏á ‡πÜ ‡πÑ‡∏î‡πâ‡πÄ‡∏•‡∏¢:
df['ingredients'] = ingredients_merged

# (‡∏≠‡∏≠‡∏õ‡∏ä‡∏±‡∏ô) ‡∏ñ‡πâ‡∏≤‡∏Ñ‡∏∏‡∏ì‡∏ï‡πâ‡∏≠‡∏á‡∏Å‡∏≤‡∏£ ‚Äú‡∏ó‡∏¥‡πâ‡∏á‡πÅ‡∏ñ‡∏ß‡∏ó‡∏µ‡πà‡∏ô‡πà‡∏≤‡∏™‡∏á‡∏™‡∏±‡∏¢‚Äù ‡∏≠‡∏≠‡∏Å‡πÑ‡∏õ‡πÄ‡∏•‡∏¢:
# if 'invalid_ing' in df.columns:
#     df = df[~df['invalid_ing']].copy()         # ‡πÄ‡∏Å‡πá‡∏ö‡πÄ‡∏â‡∏û‡∏≤‡∏∞‡πÅ‡∏ñ‡∏ß‡∏õ‡∏Å‡∏ï‡∏¥
#     df['ingredients'] = ingredients_merged.loc[df.index]

# 4) (‡∏≠‡∏≠‡∏õ‡∏ä‡∏±‡∏ô) ‡∏•‡∏ö‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡∏ä‡πà‡∏ß‡∏¢‡∏ó‡∏µ‡πà‡πÑ‡∏°‡πà‡∏ï‡πâ‡∏≠‡∏á‡∏Å‡∏≤‡∏£‡πÄ‡∏Å‡πá‡∏ö
cols_to_drop = [
    'has_dup_ing', 'dup_list', 'ingredients_list',
    'dup_after_clean', 'ingredients_cleaned',
    'invalid_reason', 'invalid_ing',
    'rank_over_5'  # ‡∏Ñ‡∏∏‡∏ì‡∏ö‡∏≠‡∏Å‡∏ß‡πà‡∏≤‡πÅ‡∏Ñ‡πà‡πÄ‡∏ä‡πá‡∏Å ‡πÑ‡∏°‡πà‡∏à‡∏≥‡πÄ‡∏õ‡πá‡∏ô‡∏ï‡πâ‡∏≠‡∏á‡πÄ‡∏Å‡πá‡∏ö‡∏Å‡πá‡∏•‡∏ö‡πÑ‡∏î‡πâ
]
df = df.drop(columns=[c for c in cols_to_drop if c in df.columns], errors='ignore')
```
