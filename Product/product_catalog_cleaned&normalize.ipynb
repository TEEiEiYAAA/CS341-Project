# Cleaning 'ingredients'

```python
# 🔹 ลบเครื่องหมาย * ทุกแถวในคอลัมน์ ingredients ก่อนตรวจ
df['ingredients'] = df['ingredients'].astype(str).str.replace('*', '', regex=False)
```

```python
# พิสูจน์ว่า ingredients มีคำซ้ำ
def has_duplicate_ingredients(text):
    parts = [p.strip().lower() for p in str(text).split(",")]
    return len(parts) != len(set(parts))

# ถ้ามีค่า True จำนวนมาก → แปลว่า “มี ingredients ซ้ำในแถว” จริง
df['has_dup_ing'] = df['ingredients'].apply(has_duplicate_ingredients)
df['has_dup_ing'].value_counts()
```

```python
# รายการแถวที่ ingredients ซ้ำ
dup_ing_df = df[df['has_dup_ing'] == True][['brand','name','ingredients']]
dup_ing_df.head(10)
```

```python
# "คำที่ซ้ำ" ในแต่ละแถว
def find_duplicates(text):
    parts = [p.strip().lower() for p in str(text).split(",")]
    dups = [p for p in parts if parts.count(p) > 1]
    return list(set(dups))

df['dup_list'] = df['ingredients'].apply(find_duplicates)
df[df['has_dup_ing'] == True][['brand','name','dup_list']].head(10)
```

```python
# “ล้างซ้ำ” ในขั้นตอน Data Cleaning
def clean_ingredients(text):
    txt = str(text).lower().replace("’", "'")
    parts = [p.strip(" .") for p in txt.split(",")]
    seen, result = set(), []
    for p in parts:
        if p and p not in seen:
            seen.add(p)
            result.append(p)
    return " | ".join(result)

df["ingredients_list"] = df["ingredients"].apply(clean_ingredients)
```

```python
def has_duplicate_after_clean(text):
    parts = [p.strip() for p in str(text).split('|')]
    return len(parts) != len(set(parts))

df['dup_after_clean'] = df['ingredients_list'].apply(has_duplicate_after_clean)
df['dup_after_clean'].value_counts()
```

```python
def extract_ingredients_only(text: str) -> str:
    s = str(text)

    # 1) รวมบรรทัดให้เป็นบรรทัดเดียว (กัน \r \n แปลก ๆ)
    s = re.sub(r'[\r\n]+', ' ', s)

    # 2) หา keyword "ingredient / ingredients" ตามด้วย :, -, – แล้วตัดส่วนหน้าออก
    #    ใช้ finditer เพื่อเลือก "ตำแหน่งสุดท้าย" เผื่อมีคำนี้ซ้ำ
    pat = re.compile(r'ingredient[s]?\s*[:\-–]\s*', flags=re.IGNORECASE)
    matches = list(pat.finditer(s))
    if matches:
        s = s[matches[-1].end():]  # ตัดทุกอย่างก่อน "ingredients:" ออก

    # 3) ทำความสะอาดสเปซ/คอมมาเล็กน้อย
    s = re.sub(r'\s*,\s*', ', ', s)   # ช่องว่างรอบคอมมาให้มาตรฐาน
    s = re.sub(r'\s+', ' ', s).strip()

    return s

# ใช้งานกับ DataFrame ของคุณ
df['ingredients_cleaned'] = df['ingredients'].apply(extract_ingredients_only)

# (ตัวเลือก) ดูจำนวนแถวที่ถูกตัด prefix ออกจริง ๆ
num_changed = (df['ingredients'] != df['ingredients_cleaned']).sum()
print(f"ตัดคำอธิบายหน้า 'Ingredients:' ออกแล้ว {num_changed} แถว")

# (ตัวเลือก) ดูตัวอย่างก่อน-หลัง
# df.loc[row_index, ['ingredients','ingredients_cleaned']]
```

```python

def check_invalid_reason(text):
    raw = str(text)
    t = raw.strip().lower()

    # รวมบรรทัดให้เป็นรูปแบบเดียว
    t_nl = re.sub(r'\r\n?|\u2028|\u2029', '\n', t)

    # 1️⃣ เคสแบบ bullet/หัวข้อ (เช่น "-Vitamin E: Acts as ...")
    bullet_lines = re.findall(
        r'^\s*[-•·]?\s*[^:\n]{1,80}:\s*[^\n,]+',  # ชื่อหัวข้อ: ตามด้วยคำอธิบาย
        t_nl,
        flags=re.MULTILINE
    )
    if len(bullet_lines) >= 1:
        return f"⚠️ ข้อความแบบบูลเล็ต/หัวข้อ {len(bullet_lines)} บรรทัดที่มี ':' (คำอธิบาย ไม่ใช่สูตร)"

    # 2️⃣ ถ้ามี '-' ที่ต้นบรรทัด (แม้ไม่มี ':' ก็ถือว่าน่าสงสัย)
    if re.search(r'^\s*-\s*[A-Za-z]', t_nl, flags=re.MULTILINE):
        return "⚠️ มีเครื่องหมาย '-' นำหน้าบรรทัด (อาจเป็นคำอธิบาย ไม่ใช่สูตรจริง)"

    # 3️⃣ ข้อความโฆษณา/บรรยายทั่วไป
    if re.search(r'visit the|provides|acts as|known to|powerful|boutique|rich in|treat|heal', t):
        return "⚠️ มีข้อความโฆษณาหรือบรรยาย (e.g., provides / acts as / known to)"

    # 4️⃣ สัญลักษณ์การตลาด
    if re.search(r'\*|\#|%|100%', t):
        return "⚠️ มีสัญลักษณ์หรือคำทางการตลาด (*, %, 100%)"

    # 5️⃣ มี ':' มากกว่า 1 ตัว (backup)
    if t.count(':') > 1:
        return "⚠️ มีเครื่องหมาย ':' มากกว่า 1 ตัว (อาจมีหลายสูตรหรือคำอธิบายใน cell เดียว)"

    # 6️⃣ สูตร Excel / #NAME?
    if raw.startswith('=') or '#name?' in t:
        return "⚠️ ข้อความมีลักษณะสูตร Excel (#NAME?, =...)"

    # 7️⃣ สั้นผิดปกติ
    if len(t.split()) < 5:
        return "⚠️ จำนวนคำ น้อยเกินไป — อาจไม่ใช่สูตร ingredients จริง"

    # 8️⃣ ตัวอักษรควบคุมหรือ newline แปลก
    if re.search(r'[\x00-\x08\x0b-\x1f]+', raw):
        return "⚠️ มีตัวอักษรควบคุมซ่อนอยู่"

    return ""
```

```python
# เพิ่มคอลัมน์ reason แสดงสาเหตุ
df['invalid_reason'] = df['ingredients'].apply(check_invalid_reason)

# flag ว่าน่าสงสัยไหม (True/False)
df['invalid_ing'] = df['invalid_reason'] != ""

# ดูตัวอย่างแถวที่น่าสงสัย
invalid_df = df[df['invalid_ing'] == True][['brand', 'name', 'ingredients', 'invalid_reason']]
invalid_df.head(10)
```

# Merge 'ingredients' column

```python
def normalize_ing_string(s: str) -> str | pd._libs.missing.NAType:
    """
    รับสตริงสูตร → คืนสตริงที่สะอาด:
      - ตัดช่องว่างเกิน
      - split ด้วย ',', strip แต่ละตัว
      - ลบว่างเปล่า
      - ลบคำซ้ำโดยคงลำดับเดิม
      - join กลับด้วย ', '
    """
    if pd.isna(s):
        return pd.NA
    s = str(s)
    # normalize space
    s = re.sub(r'\s+', ' ', s).strip()
    # tokenize
    parts = [p.strip() for p in s.split(',')]
    parts = [p for p in parts if p]  # ตัดว่าง
    # ลบซ้ำแบบคงลำดับ
    seen = set()
    dedup = []
    for p in parts:
        if p not in seen:
            seen.add(p)
            dedup.append(p)
    return ', '.join(dedup) if dedup else pd.NA

def list_to_ing_string(x) -> str | pd._libs.missing.NAType:
    """
    รับ list ของ ingredients → คืนสตริงที่สะอาด (dedup + join)
    """
    if isinstance(x, list):
        # แปลงเป็นสตริงก่อนแล้ว normalize ต่อ
        return normalize_ing_string(', '.join(map(str, x)))
    # ถ้าไม่ใช่ list (เช่น เป็น NaN หรือเป็นสตริงอยู่แล้ว) ก็ normalize แบบสตริง
    return normalize_ing_string(x)

# 1) เตรียม candidate จาก ingredients_list (ถ้ามี)
if 'ingredients_list' in df.columns:
    cand_from_list = df['ingredients_list'].apply(list_to_ing_string)
else:
    cand_from_list = pd.Series(pd.NA, index=df.index)

# 2) เตรียม candidate จาก ingredients_cleaned (ถ้ามี)
if 'ingredients_cleaned' in df.columns:
    cand_from_cleaned = df['ingredients_cleaned'].apply(normalize_ing_string)
else:
    cand_from_cleaned = pd.Series(pd.NA, index=df.index)

# 3) รวมเป็นคอลัมน์แม่ 'ingredients'
#    เลือกค่าจาก ingredients_list ก่อน (เพราะลบซ้ำแล้ว)
#    ถ้าว่างค่อย fallback ไปใช้ ingredients_cleaned
ingredients_merged = cand_from_list.fillna(cand_from_cleaned)

# (ออปชัน) ถ้าคุณต้องการ “คงข้อมูลแถวที่น่าสงสัยไว้” ก็เซ็ตตรง ๆ ได้เลย:
df['ingredients'] = ingredients_merged

# (ออปชัน) ถ้าคุณต้องการ “ทิ้งแถวที่น่าสงสัย” ออกไปเลย:
# if 'invalid_ing' in df.columns:
#     df = df[~df['invalid_ing']].copy()         # เก็บเฉพาะแถวปกติ
#     df['ingredients'] = ingredients_merged.loc[df.index]

# 4) (ออปชัน) ลบคอลัมน์ช่วยที่ไม่ต้องการเก็บ
cols_to_drop = [
    'has_dup_ing', 'dup_list', 'ingredients_list',
    'dup_after_clean', 'ingredients_cleaned',
    'invalid_reason', 'invalid_ing',
    'rank_over_5'  # คุณบอกว่าแค่เช็ก ไม่จำเป็นต้องเก็บก็ลบได้
]
df = df.drop(columns=[c for c in cols_to_drop if c in df.columns], errors='ignore')
```
